9. Generate the series of dates from 1st jan 2023 to 10th feb 2023

```python
import pandas as pd
from datetime import date,timedelta
sdate=date(2023,1,1)
edate=date(2023,2,10)
pd.date_range(sdate,edate-timedelta(days=1),freq="d")
```
